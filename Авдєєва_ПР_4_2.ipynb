Намалювати крижинку Коха з можливістю вибору рівня рекурсії.

```python
import matplotlib.pyplot as plt
import numpy as np

def koch_snowflake(order, size):
    if order == 0:
        return [(0, 0), (size, 0)]
    
    sub_segment = koch_snowflake(order - 1, size)
    new_points = []
    for i in range(len(sub_segment) - 1):
        x1, y1 = sub_segment[i]
        x2, y2 = sub_segment[i + 1]
        dx = x2 - x1
        dy = y2 - y1
        distance = np.sqrt(dx**2 + dy**2)
        unit = distance / 3
        
        # Calculate the points of the Koch curve
        pt1 = (x1, y1)
        pt2 = (x1 + dx/3, y1 + dy/3)
        pt3 = (x1 + dx/2 + np.sqrt(3) * dy/6, y1 + dy/2 - np.sqrt(3) * dx/6)
        pt4 = (x1 + 2*dx/3, y1 + 2*dy/3)
        pt5 = (x2, y2)
        
        new_points.extend([pt1, pt2, pt3, pt4])
    
    new_points.append(pt5)
    return new_points

def main():
    level = int(input("Введіть рівень рекурсії (ціле число більше або рівне 0): "))
    size = 1
    
    points = koch_snowflake(level, size)
    x, y = zip(*points)
    
    plt.figure(figsize=(8, 8))
    plt.plot(x, y)
    plt.axis('equal')
    plt.title('Koch Snowflake')
    plt.show()

if __name__ == "__main__":
    main()
```
